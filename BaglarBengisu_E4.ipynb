Generate matrices A, with random Gaussian entries, B, a Toeplitz matrix, where A ∈ R
n×m and B ∈ R
m×m,
for n = 200, m = 500.


```python
import numpy as np
import scipy.linalg as sc
A = np.random.normal(0,1,(200,500))
B = sc.toeplitz(np.random.random(500))

print("A: " , A)
print ("B: " , B)
```

Exercise 9.1: Matrix operations

Calculate A + A, AA>, A>A and AB. Write a function that computes A(B − λI) for any λ.

```python
def computation(A,B,arg) :
    return np.dot(A, B-np.dot(arg, np.eye(np.size(B,0),np.size(B,1))))

addA = A + A
print("A + A = " , addA)

A_A_T = np.dot(A,A.T)
print("A_A_T = " ,A_A_T)

A_T_A = np.dot(A.T,A)
print("A_T_A = " ,A_T_A)

AB = np.dot(A,B)
print("AB = ",AB)

arg = float(input("Enter λ "))
print(computation(A,B,arg))
```

Exercise 9.2: Solving a linear system

Generate a vector b with m entries and solve Bx = b

```python
b = np.random.random(500)
res = np.linalg.solve(B, b)
print("B * x = b" ,res)
np.allclose(np.dot(B, res), b)
```

Exercise 9.3: Norms
    
Compute the Frobenius norm of A: ||A||F and the infinity norm of B: ||B||∞. Also find the largest and
smallest singular values of B.


```python
x = np.linalg.norm(A, ord ='fro')
print("Frobenius norm of A: ", x)
y = np.linalg.norm(B, ord = np.inf)
print("Infinity form of B: ",y)
smallest = min(np.linalg.svd(B,True,False))
print("Smallest singular value of B: ",smallest)
largest = max(np.linalg.svd(B,True,False))
print("Largest singular value of B: ",largest)
```

Exercise 9.4: Power iteration
    
Generate a matrix Z, n × n, with Gaussian entries, and use the power iteration to find the largest
eigenvalue and corresponding eigenvector of Z. How many iterations are needed till convergence?
Optional: use the time.clock() method to compare computation time when varying n.

```python
import time

Z = np.random.normal(0,1,(200,200))
print(time.process_time())
v,u = np.linalg.eig(Z)
print(time.process_time())

print("Largest Eigenvalue: ", v.max())
print("The Eigenvector: ", u)
```

Exercise 9.6: Nearest neighbor
    
Write a function that takes a value z and an array A and finds the element in A that is closest to z. The
function should return the closest value, not index.
Hint: Use the built-in functionality of Numpy rather than writing code to find this value manually. In
particular, use brackets and argmin.


```python
def nearest_neighbor(z,A) :
    return A[np.fabs(A - z).argmin()]
A = np.random.random(10) * 10
print(nearest_neighbor(3,A))
```
